Write a function which will segment and extract the text overlay "Bart
**bold text**& Homer's EXCELLENT Adventure" in provided image as input. Output
 will be an Image with only the overlay visible and everything else white

Loading Required Libraries

```
import cv2
import numpy as np
```

 #Function which will segment and extract the text overlay "Bart & Homer's EXCELLENT Adventure" 

```
def getTextOverlay(input_image):
    output = np.zeros(input_image.shape, dtype=np.uint8)
    img_gray = cv2.cvtColor(input_image, cv2.COLOR_BGR2GRAY)
    _,img_thresh = cv2.threshold(img_gray,5,255,cv2.THRESH_BINARY)
    denoise1 = cv2.medianBlur(img_thresh,7) 
    output = cv2.GaussianBlur(denoise1,(5,5),0)
    return output
    
```

# The output is an image with only the overlay visible and everything else (background) white

```
if __name__ == '__main__':
    input_image = cv2.imread('/content/drive/MyDrive/Coding Test/Computer Vision/Task-1b.png')
    output = getTextOverlay(input_image)
    output_image=cv2.imwrite('/content/drive/MyDrive/Coding Test/Computer Vision/Task1.jpg', output)
```
